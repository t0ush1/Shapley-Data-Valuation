```python
import matplotlib.pylab as plt

plt.style.use("ggplot")
plt.rcParams['axes.prop_cycle']
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["mathtext.fontset"] = "custom"
plt.rcParams["mathtext.rm"] = "Times New Roman"

labels = ["Perm-Shapley", "MC-Shapley", "DIG-FL", "Extended-GTB", "CC-Shapley", "Extended-TMC", "GTG-Shapley", "OR", "$\lambda$-MR", "IPSS"]
colors = ["#FFFF00", "#8EBA42", "#8EBA42", "#FFB5B8", "#E24A33", "#348ABD", "#1F77b4", "#FF7F0E", "#2CA02C", "#D62728"]
hatchs = ["", "", "xx", "..", "//", "\\\\", "//", "**", "++", "oo"]
n = len(labels)
```

```python
from matplotlib.patches import Patch

fig, ax = plt.subplots(figsize=(10, 0.5))

# 移除坐标轴
ax.axis("off")

# 创建图例句柄
handles = [Patch(facecolor=colors[i], edgecolor="black", hatch=hatchs[i], label=labels[i]) for i in range(n)]

# 添加图例
legend = ax.legend(handles=handles, loc="center", ncol=n, frameon=True, fontsize=20)

# 调整布局并显示
# plt.tight_layout()
plt.show()
```
